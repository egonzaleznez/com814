```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np # to use numpy arrays instead of lists
import pandas as pd # DataFrame (table)
import matplotlib.pyplot as plt # to plot
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from collections import Counter
from sklearn.feature_extraction import text
```

```python
dataSet = pd.read_csv("gdrive/MyDrive/Colab Notebooks/Amazon_Unlocked_Mobile.zip", header=0, compression='zip')
print(dataSet)
```

```python
dataSet.dropna(inplace=True)
dataSet = dataSet.iloc[0:25000,:]

misDatos = dataSet[['Reviews']]
print(misDatos.head(20))
```

```python
vectorizer = TfidfVectorizer(stop_words='english',
                             min_df = 5,
                             max_df = 0.8,
                             sublinear_tf = True,
                             use_idf = True)
data_vectors = vectorizer.fit_transform(misDatos['Reviews'])

print(data_vectors)
```

```python
num_clusters = 4
model = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=350, random_state=0)
model.fit(data_vectors)
```

```python
misDatos['Clusters'] = model.labels_.tolist()
print(misDatos.head(20))
```

```python
stop_words = set(text.ENGLISH_STOP_WORDS)
cluster_word_counts = {}

for cluster_num in misDatos['Clusters'].unique():
    cluster_data = misDatos[misDatos['Clusters'] == cluster_num]
    all_words = ' '.join(cluster_data['Reviews'].tolist()).split()
    filtered_words = [word.lower() for word in all_words if word.lower() not in stop_words]
    word_counts = Counter(filtered_words)
    print(word_counts)
    print(word_counts.most_common()[0][0])
    cluster_word_counts[cluster_num] = word_counts.most_common()[0][0], word_counts.most_common()[1][0], \
                                       word_counts.most_common()[2][0], word_counts.most_common()[3][0], word_counts.most_common()[4][0]

print()
print("Most frequent word in each cluster:", cluster_word_counts)
```

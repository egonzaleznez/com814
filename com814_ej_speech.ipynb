```python
doc = open("gdrive/MyDrive/Colab Notebooks/lincoln_speech.txt", "r", encoding='latin-1')
raw_text = doc.read()
doc.close()
print(raw_text)
```

```python
text = [word for word in raw_text.split(' ') if word.isalnum()]
```

```python
sw_english = set(nltk.corpus.stopwords.words('english'))
text = [word for word in text if word not in sw_english]
```

```python
stemmer = SnowballStemmer('english')
stem_text = [stemmer.stem(w) for w in text]
stem_text
```

```python
# Lemmatization using NLTK
lemmatizer = WordNetLemmatizer()
lemm_text = [lemmatizer.lemmatize(w) for w in text]
lemm_text
```

```python
# find the first 10 words of the set
list(set(text))[:10]
```

```python
# find a frequency distribution of the words
dist = FreqDist(text)
dist
```

```python
# provide the 10 most used words
used_words = dist.most_common(10)
used_words
```

```python
# provide the words and is frequency
used_words = dist.keys()
freq_words = dist.values()
print(list(used_words))
print(list(freq_words))
```

```python
# find most used words based on a criteria
mostCommon = [w for w in used_words if len(w) > 3 and dist[w] > 2]
mostCommon
```
